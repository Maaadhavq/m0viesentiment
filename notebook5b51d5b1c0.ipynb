```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

reviews = pd.read_csv('/kaggle/input/dadaaaa/IMDB Dataset.csv') 

re = reviews['review'].values
se = reviews['sentiment'].values
trs, tess, trr, tesr = train_test_split(re, se, test_size=0.2, random_state=30)

tfidf = TfidfVectorizer(stop_words='english', max_df=0.7)
trr_tfidf = tfidf.fit_transform(trr)
tesr_tfidf = tfidf.transform(tesr)

classifier = MultinomialNB()
classifier.fit(trr_tfidf, trs)


sentiments_pred = classifier.predict(tesr_tfidf)


print("Accuracy:", accuracy_score(tess, sentiments_pred))
print("Confusion Matrix:\n", confusion_matrix(tess, sentiments_pred))
print("Classification Report:\n", classification_report(tess, sentiments_pred))
```
